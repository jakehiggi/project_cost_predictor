```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/project_data.csv')
```

```python
df.describe()
```

```python
df.head()
```

```python
numerical_features = ['Initial_Budget', 'Actual_Cost', 'Sustainability_Score', 'Resource_Usage_Index']

df[numerical_features].hist(figsize=(12, 10), bins=20, grid=False)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Industry', y='Actual_Cost', data=df)
plt.title('Actual Costs Distribution by Industry')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df[numerical_features].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Between Features')
plt.show()
```
